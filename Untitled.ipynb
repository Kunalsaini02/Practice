- there is one more operation : **eval**
- eval convert integer to int type and decimal to float type

```python
 int('10') # 10
int('10.5') # error
```

**Question-1: Average of numbers**

```python
 n1 = eval(input('Enter the number1 here: '))
 n2 = eval(input('Enter the number2 here: '))
 n3 = eval(input('Enter the number3 here: '))

average = round((n1+n2+n3)/3, 3)
print(f"The average of {n1}, {n2}, {n3} is : {average}")
```

**Question-2: Area of circle**

```python
pi = eval(input('Enter the value of pi: '))
radius = eval(input('Enter the radius of the circle: '))

area_of_circle = pi*radius*radius
print(f"The area of the circle with radius {radius} is: {area_of_circle}")
```

**Question-3: area of triangle**

```python
breadth = eval(input('Enter the breadth of triangle: '))
height = eval(input('Enter the height of triangle: '))

area_of_triangle = 0.5*breadth*height
print(f"The area of the triangle with height: {height} and breadth: {breadth} is: {area_of_triangle}")
```

**Question-4: mathematical calculations**

```python
 n1 = eval(input('Enter the number1 here: '))
 n2 = eval(input('Enter the number2 here: '))

addition = n1+n2
subtraction = n1-n2
multiplication = n1*n2
division = round(n1/n2, 3)

print(f"The addition of {n1} and {n2} is : {addition}")
print(f"The subtraction of {n1} and {n2} is : {subtraction}")
print(f"The multiplication of {n1} and {n2} is : {multiplication}")
print(f"The division of {n1} and {n2} is : {division}")
```

**Question-5: Calculate the total amount**

```python
bill_amount = eval(input('Enter the bill amount: '))
tip_amount = eval(input('Enter the tip amount: '))

total_amount = bill_amount+tip_amount
print(f"The total amount with bill value-{bill_amount} and tip value-{tip_amount} is: {total_amount}")
```

**Question-6: Calculate tax amount**

```python
Net_Salary = eval(input('Enter your salary here: '))
Tax_percentage = eval(input('Enter percentage (No need to enter % in the end):'))

Tax_amount = Tax_percentage*Net_Salary/100
print(f"The total tax amount of Rs.{Net_Salary}  at {Tax_percentage}% tax is : {Tax_amount}")
```

**Question-7: Convert Currency**

```python
Dollars = eval(input('Enter no. of Dollars: '))
INR =  eval(input('Enter the current value of one Dollar in INR: '))
INR_Currency = Dollars*INR
print(f"The ${Dollars} in INR are: Rs.{INR_Currency}.")
```
