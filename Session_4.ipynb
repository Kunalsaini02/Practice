**Triple Quotes**

- in jupyter notebook we have a **markdown** option to write an information.
- **markdown** means we are conveying some information to the user.
- But in other platforms like VS Code or pycharm, we don't have any markdown option.
- We can convey the information using triple quotes to the user, so this also called as **Doc String**.
- The code part we can provide single or double quotes.
- But conveying information about the code is triple quotes. **Triple quotes** is not for just coding, it is also used for conveying the information.
- triple quotes are used in prompt engineering.

**Boolean**

    **True**
    **False**
- True means - 1
- False means - 0
- You can think of another way **True** means **ON** and **False** means **OFF**.
