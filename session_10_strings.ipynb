**In vs Range**

- if you want to directly use the elements directly use **in** operator
    - for i in 'python' where i is characters
- if we want to access the elements using range apply index.
    - for i in range(len('python')) where i is a number


- Enumerate is an inbuilt function which give the index and element 
- in below i become index and j become element

```python
s1 = 'python'

for i,j in enumerate(s1):
    print(i,j)
```

```python
for i in s1:
    print(i)
    
    
```

```python
for i in range(len(s1)):
    print(i, s1[i])
```

**Write a program to find number of "a" in the provided string**

**Write a program to find number of "a" in the provided string**

```python
s2 = 'ola ola ola'
count = 0
print("The count of letter 'a' is:")
for i in range(len(s2)):
   
    if(s2[i]=='a'):
        count=count+1 
print(count) 
```

**Write a program to print sum of the indexes of the character "a" in the provided string**

```python
s2 = 'ola ola ola'
count = 0
print("The sum of indices of letter 'a' is:")
for i in range(len(s2)):
   
    if(s2[i]=='a'):
        count= count+i 
print(count)
```

**Write a program to print all the vowels present in the string provided**

```python
s2 = 'hai how are you'
vowels = 'aeiou'
print("The vowels in the string are:")
for i in s2:
    if i in vowels:
        print(i, end=' ')
```

**Write a program to print all the vowels present in the string provided using OR operator**

```python
s2 = 'hai how are you'
print("The vowels in the string are:")
for i in s2:
    if i=='a' or i=='e' or i=='i' or i=='o' or i=='u':
        print(i)
```

**Write a program to print all the vowels present in the string provided using OR operator but not repeated**

```python
s2 = 'hai how are you'
vowels = 'aeiou'
result = ''
print("The vowels in the string are:")
for i in s2:
    if i in vowels:
        if i not in result:
            result += i + " "      #result = result+i
print(result)
```

**Mutable vs Immutable**

- Mutbale: Chracters in a string can be changed by index operation
- Immutable: Characters in a string cannot be changed by using index operation.

```python
s2 = 'welcome'
s2[2] = 'L'
s2
```
