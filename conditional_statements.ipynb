**Conditional Statements**

- Conditional statements are used to execute the code on the basis of some condition.
- Example
    - a person greet some person if he/she know him/her otherwise no communication.
    - print('Hii') if you know me else print('Bye')


```python
today = 'rain'
movie = 'good'
# today = 'not_rain'
# movie = 'not_good'


if(today=='rain'):
    if(movie=='good'):
        print('I will go to movie.')
    else:
        print('I will go to restraunt.')
else:
    print('I will go to institute.')
```
