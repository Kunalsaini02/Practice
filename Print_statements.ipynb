```python
pi = 3.14
r = 100
area_of_circle = pi*r*r

print(f"The area of the circle having radius {r}cm is {area_of_circle} sq.cm")
print("The area of the circle having radius {}cm is {} sq.cm".format(r, area_of_circle))
```

```python
bill_amount = 2138
tip_amount = 122
total_amount = bill_amount+tip_amount

print(f"The bill amount is Rs.{bill_amount} and the tip amount is Rs.{tip_amount}. Both adds up to make the total amount to Rs.{total_amount}.")
print("The bill amount is Rs.{} and the tip amount is Rs.{}. Both adds up to make the total amount to Rs.{}.".format(bill_amount,tip_amount,total_amount))
```

```python
length = 46
breadth = 73
area_of_triangle = 0.5*length*breadth

print(f"The area of the triangle having length {length} and breadth {breadth} is {area_of_triangle}cm.")
print("The area of the triangle having length {} and breadth {} is {}cm.".format(length,breadth,area_of_triangle))
```

```python
number1 = 7392
number2 = 3883

addition = number1+number2
subtraction = number1-number2
multiplication = number1*number2
division = number1/number2

print(f"The addition of {number1} and {number2} is {addition}.")
print(f"The subtraction of {number1} and {number2} is {subtraction}.")
print(f"The multiplication of {number1} and {number2} is {multiplication}.")
print(f"The division of {number1} and {number2} is {division}.")
```

- Another way to print using .format() function

```python
print("The addition of {} and {} is {}.".format(number1,number2,addition))
print("The subtraction of {} and {} is {}.".format(number1,number2,subtraction))
print("The multiplication of {} and {} is {}.".format(number1,number2,multiplication))
print("The division of {} and {} is {}.".format(number1,number2,division))
```
